```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set_style('darkgrid',{'axes.grid':True})
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='Dark2')
plt.show
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
plt.show
```

## UNIVALIATE (one variable)

```python
#strip plot
sns.stripplot(y='tip',data=tips,jitter=True)
plt.ylabel('tip($)')
plt.show
```

```python
#grouping with stripplot
sns.stripplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=False)
plt.show()
```

```python
#swarm plot
#More denser
sns.swarmplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.show
```

```python
#changing orientation
sns.swarmplot(x='tip',y='day',data=tips,hue='sex',orient='h')
plt.show
```

## Box and violin plot

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip')
```

```python
plt.subplot(1,2,2)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip')
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip')
plt.subplot(1,2,2)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip')
plt.show()
```

```python
#combining plot
sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgrey')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.show()
```

## Bivariate(Two variable)

```python
#joint plot
#it uses 2 graphs(scattered and histogram)
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#density plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

```python
#Pair plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
#corelation
tips.corr()
```

```python
#covariance heat map
sns.heatmap(tips.corr(),vmin=-1,vmax=1,cmap='ocean')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Garbage1.csv')
```

```python
print(data.dtypes)
```

```python
f,ax=plt.subplots(figsize=(11,5))
sns.boxplot(x='binno',y='full_time',data=data)
```

```python
f,ax=plt.subplots(figsize=(7,3))
sns.barplot(x='binno',y='full_time',data=data)
```

```python
f,ax=plt.subplots(figsize=(11,5))
sns.boxplot(x='binno',y='full_time',hue='lat',data=data)
```

```python
sns.barplot(x='binno',y='full_time',data=data)
```

```python
su=data.loc[data['full_time']==1,'lat'].dropna()
sns.distplot(su)
```

```python
sns.pairplot(data.dropna())
```

```python
grid=sns.FacetGrid(data,col='binno',row='lat',size=2.6,aspect=1.6)
grid.map(plt.hist,'full_time',alpha=.5,bins=20)
grid.add_legend()
```
